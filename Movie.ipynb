Task 2: Data Analysis

1.	Divide the 'gross' and 'budget' columns by 1000000 to convert '$' to 'million $'

```python
Movie["Gross"] = Movie["Gross"]/1000000
Movie
```

```python
Movie["budget"] = Movie["budget"]/1000000
Movie
```

2.	Create the new column named 'profit' by subtracting the 'budget' column from the 'gross' column

```python
Movie["Profit"] = Movie["Gross"] - Movie["budget"]
Movie
```

3.	Sort the data frame with the 'profit' column as reference using the 'sort_values' function. Make sure to set the argument 'ascending' to 'False'

```python
sorted_one = Movie.sort_values(by="Profit", ascending=False)
sorted_one
```

4.	Get the top 10 profitable movies by using position-based indexing. Specify the rows till 10 (0-9)

```python
top10 = sorted_one.iloc[0:10]
top10
```

5.	Find the movies with negative profit

```python
negative = Movie[Movie["Profit"]<0]
negative
```

6.	Change the scale of MetaCritic(MetaCritic score is on a scale of 100 whereas the IMDb_rating is on a scale of 10. First convert the MetaCritic column to a scale of 10.)

```python
Movie["MetaCritic"] = Movie["MetaCritic"] / 10
Movie["MetaCritic"] 
```

7.	Find the average ratings

```python
Movie["Average Rating"] = (Movie["IMDb_rating"] + Movie["MetaCritic"]) / 2
Movie
```

8.	Sort in descending order of average rating

```python
Descending_Order = Movie.sort_values (by="Average Rating", ascending=False)
Descending_Order
```

9.	Find the movies with metacritic-rating < 0.5 and also with the average rating of >8

```python
Find_Movies = Movie[(Movie["MetaCritic"] >0.5) & (Movie["Average Rating"] > 8)]
Find_Movies
```

10.	The data frame has three columns to help you out for the same, viz. actor_1_facebook_likes, actor_2_facebook_likes, and actor_3_facebook_likes. Your objective is to find the trios which has the most number of Facebook likes combined. That is, the sum of actor_1_facebook_likes, actor_2_facebook_likes and actor_3_facebook_likes should be maximum. Find out the top 5 popular trios, and output their names in a list.

```python
Movie["Total Likes"] = (Movie["actor_1_facebook_likes"] + Movie["actor_2_facebook_likes"] + Movie["actor_3_facebook_likes"])
Movie["Total Likes"]
```

```python
top_trios = Movie.sort_values(by="Total Likes", ascending=False)
top_trios
```

```python
top5_Trios = top_trios.iloc[0:5]
top5_Trios
```

```python
name_list = top5_Trios[["actor_1_name"] + ["actor_2_name"] + ["actor_3_name"]].values
print(name_list)
```
